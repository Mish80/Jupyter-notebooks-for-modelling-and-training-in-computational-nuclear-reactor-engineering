# Introduction: The Importance of Prompt Fission Neutrons <a id="intro"></a>

Approximately 99 % of the neutrons released during fission are released almost right after fission occurs, about $10^{-14}$ s. These neutrons are known as prompt fission neutrons (PFN)s, as opposed to delayed fission neutrons (DFN) which are emitted somewhat later on in the fission process, from a few seconds to a few minutes [@lamarshIntroductionNuclearEngineering2001].



**Figure 1: Fission Event [@PromptNeutronsDefinition]**

As a result of both elastic and inelastic collisions with nuclei, the prompt fission neutrons slow down after they are emitted. While thermal reactors rely on moderators to slow down neutrons and optimize fission with $\text{U}^{235}$ at thermal energies (0.01 eV to 1 eV), fast reactors take advantage of fast neutrons to fission certain isotopes like $\text{Pu}^{239}$. Naturally, prompt neutrons, if not absorbed or used to fission, may escape from the reactor core. When creating a biological shield around a reactor core, the radiation as a result of the escaped prompt neutrons, need to be taken into account the most [@lamarshIntroductionNuclearEngineering2001].

The energy spectrum of PFNs is crucial for various applications in the nuclear field due to their significant role in, for example:<br>

* Criticality Calculations: By analyzing PFNs, one may ascertain the quantity of neutrons generated for each fission event. This information is crucial for determining criticality, which is a prerequisite for a reactor's ability to maintain a nuclear chain reaction.<br>
* Reactor kinetics: Modeling a reactor's behavior under a variety of circumstances, including short-term scenarios that impact power and safety, requires an understanding of the energy distribution of neutrons released.<br>
* Fuel Depletion and Breeding: To optimize fuel consumption and create advanced fuel cycles, fuel depletion and breeding processes must be modeled using PFN data.<br>
* Neutron Transmutation: Predicting the neutron-induced transmutation of different materials within a reactor core, such as fuel cladding and structural components, requires an understanding of the PFNs.<br>

Hence, the type of reactor and its operating parameters are largely dependent on the average energy of neutrons emitted during fission [@lamarshIntroductionNuclearEngineering2001].

Another important factor is the average fission neutron multiplicity or yield, $\bar{\nu}$, where $\bar{\nu}={\bar{\nu}}_p+\ {\bar{\nu}}_d$, which depends on both the fissioning nucleus and the incident neutron energy $\text{E}_n$. 

**Table 1: Total Fission Neutrons per Fission, $\bar{\nu}$, for several important nuclides [@shultisFundamentalsNuclearScience2002].**

| Nuclide | Fast Fission ( $\bar{\nu}$ ) | Thermal Fission ( $\bar{\nu}$ )|
|---|:---|:---|
| $\text{U}^{235}$| 2.57 | 2.43 |
| $\text{U}^{238}$ | 3.09 | 2.87 |
| $\text{Pu}^{239}$ | 2.79 | - |

It can be noted that the yield changes only slightly for fast & thermal fission.

In a PFN energy spectrum, the energy of the released neutrons, typically expressed in MeV, is usually shown against the number of neutrons released per unit energy. The distribution of PFNs across various energy ranges and the average energy of PFNs can be determined by analysing the spectrum. How to attain such a model will be analysed in the next section. # Mathematical Model Approach <a id="model"></a>

The complexity of modelling the energy spectrum of PFNs arises from the intricate details of the fission process and the influence of various factors, for example: <br>

* Statistical nature: Fission is a statistical process in which a heavy nucleus splits into lighter fragments and frequently releases several neutrons in the process. These neutrons have a probabilistic distribution in terms of both quantity and energy, rather than being constant.
* Residual nuclear excitation: Following fission, the resultant pieces release a variety of particles, including neutrons, as a way of losing energy. The excitation energy distribution of the pieces determines the energy distribution of these released neutrons.
* Coulomb repulsion: During fission, the charged pieces resist one another, which has an impact on the energy available for neutron emission.<br>
  
Despite its complexity, the single-parameter Maxwell-Boltzmann distribution, or simply "Maxwellian," is the earliest and most basic model of the PFNs and is still used in some applications today. It is dependent on a temperature parameter, T, expressed in MeV, which represents the nuclear temperature at the moment of neutron emission, and E, expressed in MeV, which is the energy of the neutron that is emitted [@losalamosnationallaboratoryPromptFissionNeutron2016]:
$$ \chi_M\left(E,T\right)=\frac{2\sqrt E}{\sqrt{\pi T^3}}\ e^{(-E/T)} $$

where $\chi_M$ represents the number of prompt neutrons emitted with energy E. 

However, the Maxwellian model ignores the following [@losalamosnationallaboratoryPromptFissionNeutron2016]:<br>

* Fragment Motion: When a nucleus splits (fissions) and releases neutrons, the motion and energy of the precursor and fission fragments are not taken into consideration. 
* Movement of Entire System: Because of not considering the motion of the precursor and fission fragments, the models predicted distribution of neutrons becomes oversimplified. <br>
* 
As such, the Maxwellian model lacks predictive ability.

The following modified two-parameter Maxwellian distribution, or the Watt spectrum, namely improves upon the following [@losalamosnationallaboratoryPromptFissionNeutron2016]):<br>

* Fragment Motion: Neutron emission from an average fission fragment with average kinetic energy Ef, expressed in MeV, is considered.
* Transformation to Real-World Reference: Adjusting the calculations to account for the movement of the entire system in the real world, not just the movement of the neutron.<br>
  
Hence, the Watt spectrum is a more detailed model:
$$\chi_W\left(E,T,E_f\right)=\frac{1}{\sqrt{\pi T\ E_f}}\sinh\sqrt{\frac{4E_fE}{T^2}}\ e^{\left[-\frac{E_f+E}{T}\right]}=\frac{e^{\left(\frac{-E_f}{T}\right)}}{\sqrt{\pi T E_f}}\sinh{\sqrt{\frac{4E_fE}{T^2}}}e^{\left[-\frac{E}{T}\right]}$$

Simplified further into the following equation to accommodate other nuclides:
$$\text{S}_{PFN}\left(E\right)=a\ e^{\left(\frac{-E}{b}\right)}\sinh{\sqrt{c\ E}}$$

Where:
$$a=\frac{e^{\left(\frac{-E_f}{T}\right)}}{\sqrt{\pi T\ E_f}}\ ;b=T;c=\frac{4E_f}{T^2};\ \bar{E}=\frac{3}{2}T+E_f\ $$

**Table 2: Parameters for the Watt Approximations [@shultisFundamentalsNuclearScience2002]**

| Nuclide | Type of Fission | T | $\text{E}_f$ | a | b | c |
|---|---|:---|:---|:---|:---|:---|
| $\text{U}^{235}$| Thermal | 1.035 | 0.4340 | 0.5535 | 1.0347 | 1.6214 |
| $\text{U}^{238}$ | Fast | 1.027 | 0.4159 | 0.5759 | 1.0269 | 1.5776 |
| $\text{Pu}^{239}$ | Thermal | 1.159 | 0.4159 | 0.5759 | 1.0269 | 1.5776 |

## Declaration of Constants

Parameters for the Watt Approximation from Table 2 need to be declared as constants.

```python
a_U235 = 0.35535 
b_U235 = 1.0347
c_U235 = 1.6214

a_U238 = 0.4159 
b_U238 = 1.0269
c_U238 = 1.5776

a_Pu239 = 0.5710 
b_Pu239 = 1.1593
c_Pu239 = 1.2292
```

## Energy Range Set-up

The expected upper and lower PFN energy level range needs to be declared such that the appropriate graph can be plotted. Around 70-80% of PFNs are emitted with the energy between 0-10 MeV [@lamarshIntroductionNuclearEngineering2001]. One can then safely set the lower and upper bounds to 0 and 15 MeV respectively.

```python
# We will use a value slightly above machine epsilon as the lower bound
# (smallest positive number representable by your computer). 
# This helps avoid potential issues with extremely small values.
E_min = np.log10(1.00000001E-11)
E_max = np.log10(15)
E = np.logspace(E_min, E_max, num=9999, endpoint=True, base=10.0, dtype=float, axis=0)
# Logspace will be used instead of linspace for now - incase we decide to create a log graph
```

## Defining the Watt Energy Spectrum Equation

```python
# Equation for the Watt Spectrum
def S_pfn_U235(E):
    return a_U235*np.exp(-E/b_U235) * np.sinh(np.sqrt(c_U235*E))
def S_pfn_U238(E):
    return a_U238*np.exp(-E/b_U238) * np.sinh(np.sqrt(c_U238*E))
def S_pfn_Pu239(E):
    return a_Pu239*np.exp(-E/b_Pu239) * np.sinh(np.sqrt(c_Pu239*E))
```

## Graph Plotting - Linear scale

The first graph to be plotted is expected to look like Figure 2, a linearly scaled graph. 

```python
# Linear plot - expect the same output as seen in Figure 2
fig, ax = plt.subplots(1,1)
ax.plot(E, S_pfn_U235(E), label=r'Spectrum for PFN emission by ${_{\;\;92}^{235}}\mathsf{U}$',  color='Green', linestyle='-', marker=None, markersize=10)
ax.plot(E, S_pfn_U238(E), label=r'Spectrum for PFN emission by ${_{\;\;92}^{238}}\mathsf{U}$', color='Blue', linestyle='-', marker=None, markersize=10)
ax.plot(E, S_pfn_Pu239(E), label=r'Spectrum for PFN emission by ${_{\;\;94}^{239}}\mathsf{Pu}$', color='Violet', linestyle='-', marker=None, markersize=10)

ax.set_title('Prompt Fission Neutorn Energy Spectrum', fontsize=18)
ax.set_xlabel('Neutron energy (MeV)', fontsize=14)
ax.set_ylabel('PFN Yield', fontsize=14)
ax.set_xscale('linear')
ax.set_yscale('linear')
ax.set_xlim(0, 15)
ax.set_ylim(0, 0.4)

ax.fill_between(E, 0, S_pfn_U235(E), where=(E >= 1) & (E <= 3), color='gray', alpha=0.3)

# loc=0: Sets the legend's location in the plot area. In this case, 0 represents the upper right corner (other options exist for different placements).
ax.legend(loc=0, fontsize=12)
# enables minor ticks on the axes for better readability
ax.minorticks_on()
ax.grid(axis='both', color='gray', alpha=1.0, linestyle=':', linewidth=1.0)
# adjust the spacing between the plot elements and axes for a cleaner presentation.
fig.tight_layout(pad=4, w_pad=2, h_pad=2)
```

## Calculating the Average no. of PFNs Emitted in a Certain Energy Range

```python
# Calculating the average no. of PFNs emitted in a certain energy range
E1=1
E2=3
# The below code is placed in the above code block for functional purposes
# ax.fill_between(E, 0, S_pfn_U235(E), where=(E >= E1) & (E <= E2), color='gray', alpha=0.3)
TotalYield = sum(S_pfn_U235(E))
Yield = sum(S_pfn_U235(E[(E >= E1) & (E <= E2)] ))
PercentYield = (Yield/TotalYield)*100
print('Percentage of U-235 fission neutrons emitted between',E1,'MeV and', E2, 'MeV is',PercentYield, '%.')

# Another technique

# # Assuming E is your x-axis data and S_pfn_U235 is the yield data for U235 (used for shading)
# dE = E[1] - E[0]  # Assuming constant spacing between energy points
# # Integral of shaded area
# integral_shaded = np.trapz(S_pfn_U235(E)[(E >= E1) & (E <= E2)], E[(E >= E1) & (E <= E2)])
# # Integral of unshaded area (assuming entire x-axis range)
# integral_unshaded = np.trapz(S_pfn_U235(E), E)
# # Calculate percentage
# percentage_shaded = (integral_shaded / (integral_shaded + integral_unshaded)) * 100
# print(f"Percentage of U-235 fission neutrons emitted between the bounds is: {percentage_shaded:.2f} %")
```

## Calculating the Average Energy of PFNs Emitted 

```python
# Calculating the average energy of PFNs emitted 
average_energy = sum(S_pfn_U235(E))/sum(E)
print('The average energy of the PFNs for U-235 is =', average_energy,'MeV')
```

## Calculating the Average Yield of PFNs Emitted 

```python
average_yield = statistics.mean(S_pfn_U235(E))
print('The average yield of PFNs for U-235 is =', average_yield, 'probability')
```
